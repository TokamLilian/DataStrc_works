### Binary Search Tree

#### Basic Operations: 

- create(T, n, max) : Create priority queue from an array T [1, n]
- size(p) : Number of elmts in queue p
- top(p) : Most prior ekement of queue p
- deleteTop(p) : Delete most prior element (call top and remove it)
- insert(p, x) : Insert element x in the queue p
- edit(p, i, x') : Edit element i to become x (optional)

````python
n = 1#
T = [1]#
max = 2

def create(T, n, max):
#consntat
    if n>max or n<0 or max <0 : 
        return "error" 
    
def size(p):
#constant
    return n

def top(p):
#O(n)
    if n == 0 : return "error"
    n = T[1]
    for i in range(2, n) :
        if T[i] * p > n*p :   #* means the array of the queue object
            n = T[i]

    return n

def insert(p, x):
#constant
    if n >= max : return "error"
    n += 1
    T[n] = x

def deleteTop(p):
#O(n)
    if n == 0 : return "error"
    for i in range(2, n) :
        if T[i] * p < T[i-1] * p : 
            temp = T[i]
            T[i] = T[i-1]
            T[i-1] = temp
    n -= 1

def edit(P, i, x):
#constant
    if i<1 or i>n: return "error"
    T[i] = x
````
